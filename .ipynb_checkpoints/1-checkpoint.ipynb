```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

### 1.神经元模型

神经元模型对神经的膜电位变换进行建模。存在许多神经元模型。这里介绍几种重要的模型：
金字塔：**Hodgkin-Huxley**模型, 在SNN中被使用的**Leaky integrate-and-fire**模型以及**Inter-spike interval**模型。此外也对非spike-based的**发火率(firing rate)模型** 进行介绍


#### 1.1 Hodgkin-Huxley模型 (HH模型)

1.1.1 Hodgkin-Huxley模型上膜的等価回路模型


A.L. Hodgkin と A.F. Huxley在1952年使用对**電位固定法 (voltage-clamp)**进行实验得到的模型。
对于HH模型来说存在等价的电路模型，也被称为 **膜の並列等価回路モデル（parallel condutance model）**
在该电路模型中，细胞膜成为了电容，在细胞膜中存在的**离子通道(ion channel)**为可变电阻，离子通道能够选择性地令特定离子通过（例如Ca离子，Na离子）。
离子通道根据离子的种类不同存在不同的**Conductance(电导)**以及**平衡电位(equilibrium potential)**。
HH模型假定Na+ channel, K+ channel以及leak current channel.

漏电流的离子通道是当时无法确定的通道，意味着从膜中漏出电流的通道。如今已被探明的是，漏电流许多情况是由Cl-离子引起的(chloride ion)

现在开始开始使用等价的回路模型进行建模

$C_m$: 细胞膜膜容量

$I_m(t)$: 通过细胞膜的电流

$I_{Cap}(t)$: 流过膜电容的电流

$I_{Na}(t),I_{K}(t)$: 分别为通过Na, K离子通道流出的电流

$I_L(t)$: 漏电流

$$I_m(t) = I_{Cap}(t) + I_{Na}(t) + I_K(t) + I_L(t)$$

于是假定膜电位为$V(t)$, 由Kirchhoff’s Voltage Law得到

$$C_m\frac{dV(t)}{dt}=I_m(t)-I_{Na}(t)-I_K(t)-I_L(t)$$

等式左侧也为$I_{Cap}(t)$, 将三类离子通道电流项移项之后得到$I_{Cap}$, 并转换为膜电位的微分形式

Hodgkin与Huxley通过实验对Na K L电流进行了建模

$$I_{Na}(t) = g_{Na}\cdot m^3 h(V-E_{Na})$$

$$I_L(t)=g_K\cdot n^4(V-E_K)$$

$$I_L(t)=g_L(V-E_L)$$

$g_{Na}, g_K$分别为Na+, K+的最大电导

$g_L$是遵从欧姆法则的电导，$L$电导不随时间变化

$m$:Na+电导的活性化参数

$h$为Na+的不活性化参数

$n$为K+电导的活性化参数, 表示门的开闭概率

根据以上，可以了解，HH模型可由$V, m,h,n$ 4变量表示

这些变量遵从下列的x分别带入m,h,n的微分方程式。

$$
\frac{dx}{dt}=\alpha_x(V)(1-x)-\beta_x(V)x
$$

其中$V$为函数，$alpha_x(V),\beta_x(V)$由h,m,n不同而不同。满足下列6等式

$\alpha_m(V)=\frac{0.1(25-V)}{exp[(25-V)/10]-1}$

$\alpha_h(V)=0.07exp(-V/20)$

$\alpha_n(V)=\frac{0.01(10-V)}{exp[(10-V)/10]-1}$

$\beta_m(V)=4exp(-V/18)$

$\beta_h(V)=\frac{1}{exp[(30-V)/10]-1}$

$\beta_n(V)=0.125exp(-V/80)$


#### 1.1.2 HH模型的实装

$$C_m = 1.0, g_{Na} = 120, g_K = 36, g_L = 0.3,
E_{Na} = 50.0, E_K = −77, E_L = −54.387$$

```python
class HodgkinHuxleyModel:
    def __init__(self, dt=1e-3, solver="RK4"):
        self.C_m = 1.0 # 膜容量 (uF/cm^2)
        self.g_Na = 120.0 # Na+ channel 最大电导 (mS/cm^2)
        self.g_K = 36.0 # K+ channel 最大电导 (mS/cm^2)
        self.g_L = 0.3 # 漏离子通道 最大电导 (mS/cm^2)
        self.E_Na = 50.0 # Na+ 平衡電位 (mV)
        self.E_K = -77.0 # K+ 平衡電位 (mV)
        self.E_L = -54.387 # 漏离子通道 平衡電位 (mV)
        self.solver = solver # 微分方程求解器
        self.dt = dt

        # HH模型的4参数, 初始化一个状态
        # V, m, h, n
        self.states = np.array([-65, 0.05, 0.6, 0.32])
        self.I_m = None
    
    # 微分方程求解器
    def Solvers(self, func, x, dt):
        # 4th order Runge-Kutta 法
        if self.solver == "RK4":
            k1 = dt * func(x)
            k2 = dt * func(x + 0.5 * k1)
            k3 = dt * func(x + 0.5 * k2)
            k4 = dt * func(x + k3)
            return x + (k1 + 2 * k2 + 2 * k3 + k4) / 6
    
        # Euler 法
        elif self.solver == "Euler":
            return x + dt * func(x)
        else:
            return None
    
    # 关于离子门的6函数
    def alpha_m(self, V):
        return 0.1*(V+40.0)/(1.0 - np.exp(-(V+40.0) / 10.0))
    def beta_m(self, V):
        return 4.0*np.exp(-(V+65.0) / 18.0)
    def alpha_h(self, V):
        return 0.07*np.exp(-(V+65.0) / 20.0)
    def beta_h(self, V):
        return 1.0/(1.0 + np.exp(-(V+35.0) / 10.0))
    def alpha_n(self, V):
        return 0.01*(V+55.0)/(1.0 - np.exp(-(V+55.0) / 10.0))
    def beta_n(self, V):
        return 0.125*np.exp(-(V+65) / 80.0)

    def I_Na(self, V, m, h):
        return self.g_Na * m**3 * h * (V - self.E_Na)
    
    def I_K(self, V, n):
        return self.g_K * n**4 * (V - self.E_K)

    def I_L(self, V):
        return self.g_L * (V - self.E_L)


    def dALLdt(self, states):
        V, m, h, n = states
        dVdt = (self.I_m - self.I_Na(V, m, h) - self.I_K(V, n) - self.I_L(V)) / self.C_m
        dmdt = self.alpha_m(V)*(1.0-m) - self.beta_m(V)*m
        dhdt = self.alpha_h(V)*(1.0-h) - self.beta_h(V)*h
        dndt = self.alpha_n(V)*(1.0-n) - self.beta_n(V)*n
        return np.array([dVdt, dmdt, dhdt, dndt])

    def __call__(self, I):
        self.I_m = I
        # 微分函数，当前状态， dt
        # 根据各项微分的结果，以及当前状态，获取下一时间t + dt下的状态:(V,m,h,n)
        states = self.Solvers(self.dALLdt, self.states, self.dt)
        self.states = states
        return states
```

```python
dt = 0.01
T = 400 # (ms)
nt = round(T/dt) # simulation step数
time = np.arange(0.0, T, dt)

# 生成刺激電流 (uA/cm^2)
I_inj = 10 * (time > 100) - 10 * (time > 200) + 35 * (time > 250) - 35 * (time > 350)

HH_neuron = HodgkinHuxleyModel(dt=dt, solver="Euler")
X_arr = np.zeros((nt, 4)) # 記録用数组, V,

# simulation nt steps
for i in tqdm(range(nt)):
    # in each step, give HH model the stimulas current
    X = HH_neuron(I_inj[i]) # __call__
    X_arr[i] = X
```

```python
plt.figure(figsize=(5, 5))

# V
plt.subplot(3,1,1)
plt.plot(time, X_arr[:,0], color="k")
plt.ylabel('V (mV)'); 
plt.xlim(0, T)

plt.subplot(3,1,2)
plt.plot(time, I_inj, color="k")
plt.ylabel('$I_{inj}$ ($\\mu{A}/cm^2$)')
plt.xlim(0, T)


# m,h,n
plt.subplot(3,1,3)
plt.plot(time, X_arr[:,1], 'k', label='m')
plt.plot(time, X_arr[:,2], 'gray', label='h')
plt.plot(time, X_arr[:,3], 'k', linestyle="dashed", label='n')
plt.xlabel('t (ms)'); plt.ylabel('Gating Value'); plt.legend(loc="upper left")

plt.show()
```

#### 1.1.3 anodal break

电流流入神经元会导致膜电位变化，膜电位V在超过一定阈值后会发火(firing)

但是膜电位阈值是怎样的程度呢？答案是不一定。这样的现象叫做anodal break。或anode break excitation(ABE)。

ABE是一种现象，当停止施加**过极性电流**时，不仅**膜电位恢复到静止膜电位**，而且进行更进一步的**去极化后发火**。因为HH模型可以再现ABE，所以我们来确认一下是怎样的现象。这只需变更前一节安装的HH模型中的输入电流即可进行。




```python


dt = 0.01; 
T = 250 # !変更
nt = round(T/dt) 
time = np.arange(0.0, T, dt)

# 生成刺激電流 (uA/cm^2)
I_inj = -10 * (time>50) + 10 * (time>100) - 20 * (time>150) + 20 * (time>200) # !变更

HH_neuron = HodgkinHuxleyModel(dt=dt, solver="Euler")
X_arr = np.zeros((nt, 4)) 

# simulation nt steps
for i in tqdm(range(nt)):
    # in each step, give HH model the stimulas current
    X = HH_neuron(I_inj[i]) # __call__
    X_arr[i] = X
```

```python
plt.figure(figsize=(5, 5))

# V
plt.subplot(3,1,1)
plt.plot(time, X_arr[:,0], color="k")
plt.ylabel('V (mV)'); 
plt.xlim(0, T)

plt.subplot(3,1,2)
plt.plot(time, I_inj, color="k")
plt.ylabel('$I_{inj}$ ($\\mu{A}/cm^2$)')
plt.xlim(0, T)


# m,h,n
plt.subplot(3,1,3)
plt.plot(time, X_arr[:,1], 'k', label='m')
plt.plot(time, X_arr[:,2], 'gray', label='h')
plt.plot(time, X_arr[:,3], 'k', linestyle="dashed", label='n')
plt.xlabel('t (ms)'); plt.ylabel('Gating Value'); plt.legend(loc="upper left")

plt.show()
```


从**过极化的状态**返回到**静止膜电位时** (T=100,200附近)，Na+ channel被激活（Na+沟道的激活参数**m增加**，失活参数**h减少**）,膜电位去极化，Na+沟道被再次被激活，这样的积极反馈过程（**自我再生的过程**）（当然，这个过程是通常的活动电位发生的机制）。



这时，firing所需的阈值会有**随着膜电位的降低而下降**的情况。

自己推测在发火后膜电位下降。到-状态
在100左右尝试回到静止膜电位，也确实回到了，但是继续进行发火回到-50左右。

"阳极断裂激发 （ABE） 是一种电生理现象，神经元响应于超极化电流的终止而激发动作电位"
如图，虽然I回到了0，但是V在回到0+后又发火到了动作电位




当超极化电流施加到膜上时，膜上的电位下降（静息电位变为负值）, T=50附近；

在这种下降之后，动作电位所需的阈值下降（因为**阈值与跨膜电位直接相关** - 它们一起上升和下降）。


ABE在超极化电流终止后产生：**在没有超极化刺激的情况下，电位迅速上升**，但**动作电位阈值保持在较低的值**。因此，电位是超阈值的：足以在细胞内引起动作电位。

### 1.2 Leaky integrate-and-fire モデル

#### 1.2.1 LIF简单的实装

不考虑生理的举动, 在到达膜电位的阈值前对输入电流积分。这样的模型被称为 **Integrate-and-fire (IF, 積分発火)**

若是在IF模型上考虑漏电流(这样的漏电流是由于离子扩散之类的事件产生的)，则获得LIF模型

$V_m(t)$: 神经元的膜电位

$V_{rest}$: 静止膜电位

$I(t)$: 输入电流

$R_m$: 膜电阻

$\tau_m$: 膜电位的时常数 (=$R_m\cdot C_m$)

满足下列式

$$\tau_m\frac{dV_m(t)}{dt}=-(V_m(t)-V_{rest})+R_mI(t)$$

$(V_m(t)-V_{rest})$也有仅使用$V_m$的情况。右侧的RI部分也有仅使用$I$的情况。

$V_m$超过阈值$V_{th}$ (*由于下标文字由th开始，也有用$V_{\theta}$表示的情况。)之后，会引起**脱分极**, 不断上升直到Peek电位$V_{peak}$, 发火后引起再分极。降低直到$V_{rest}$以下。

发火后，一段时间内$\tau_{ref}$之间的膜电位不变化。（该时期被称为 **refactory time period**） (*根据实现的不同，也有在改期间允许电位变化但不发火的情况。)

以下开始LIF是实现，简单起见，首先不设计class, 仅对单个神经元进行建模。

#### 1.2.1.1 常数定义

需要注意的是使用的I不是指真实的I本身，而是指与其成比例的值:RI。若R=1,则为I变量表示输入电流本身

```python
dt = 5e-5; T = 0.4 # (s)
nt = round(T/dt) # Simulation steps
tref = 2e-3 # refactory time period (s)
tc_m = 1e-2 # 膜的时间常数 (s)
vrest = -60 # 静止膜電位 (mV)
vreset = -65 # 重置電位 (mV)
vthr = -40 # 阈值電位 (mV)
vpeak = 30 #　峰值電位 (mV)
t = np.arange(nt)*dt*1e3 # 時間 (ms)
I = 25*(t>50) - 25*(t>350) # 入力電流 (に比例する値)(mV) 
# 初期化
v = vreset # 膜電位初期値
tlast = 0 # 最後のスパイクの時間を記録する変数
v_arr = np.zeros(nt) # 膜電位记录数组
```

#### 1.2.1.2 主部分

```python
for i in tqdm(range(nt)): # 执行simulation step回
    dv = (vrest - v + I[i]) / tc_m # 膜電位変化量
    v = v + ((dt * i) > (tlast + tref)) * dv * dt # 更新到新状态
    s = 1 * (v >= vthr) # 発火確認
    tlast = tlast * (1 - s) + dt * i * s # 更新発火時刻, 若未发火，则不变
    v = v * (1 - s) + vpeak * s # 若发火，则更新peek电位
    v_arr[i] = v # 保存膜電位値
    v = v * (1 - s) + vreset * s # 正在发火则reset膜电位
```

```python
plt.figure(figsize=(5, 3))
plt.plot(t, v_arr)
plt.xlim(0, t.max())
plt.xlabel('Time (ms)'); plt.ylabel('Membrane potential (mV)')
plt.show()
```

### 1.2.2 LIF模型Class

之所以使用 "CurrentBasedLIF"这一名称，是因为还有另一个模型 "基于传导的 LIF"，对突触电流的描述有所不同。

```python
class CurrentBasedLIF:
    def __init__(self, N, dt=1e-4, tref=5e-3, tc_m=1e-2, vrest=-60, vreset=-60, vthr=-50, vpeak=20):
        self.N = N
        self.dt = dt
        self.tref = tref
        self.tc_m = tc_m
        self.vrest = vrest
        self.vreset = vreset
        self.vthr = vthr
        self.vpeak = vpeak
        self.v = self.vreset * np.ones(N)
        self.v_ = None
        self.tlast = 0
        self.tcount = 0
    
    def initialize_states(self, random_state=False):
        if random_state:
            self.v = self.vreset + \
            np.random.rand(self.N)*(self.vthr-self.vreset)
        else:
            self.v = self.vreset*np.ones(self.N)
        self.tlast = 0
        self.tcount = 0
    
    def __call__(self, I):
        dv = (self.vrest - self.v + I) / self.tc_m
        v = self.v + (self.dt*self.tcount>(self.tlast+self.tref))*dv*self.dt

        s = 1 * (v >= self.vthr) #发火的情况下s = 1, 其余情况为0。

        self.tlast = self.tlast * (1 - s) + self.dt * self.tcount * s # 発火時刻の更新
        v = v * (1 - s) + self.vpeak * s # 若超过阈值则将膜电位设置为峰值电位
        self.v_ = v # 発火時の電位も含めて記録するための変数
        self.v = v * (1 - s) + self.vreset * s # 発火時に膜電位をリセット
        self.tcount += 1

        return s
```

```python
neuron = CurrentBasedLIF(N=1)
v = neuron.v_
```

### 1.2.3 LIF模型的F-I曲线

#### 1.2.3.1 根据数値的計算的F-I curve描画

**frequency-current (F-I) curve**也被称为**neuronal input/output (I/O) curve**, 描述了LIF模型对于输入电流的大小，发火率的变化将会如何变化。

```python
dt = 5e-5; T = 1; nt = round(T/dt)
tref = 5e-3; # Can be adjusted to 0 to see how the curve would be changed.
tc_m = 1e-2; vrest = 0; vreset = 0; vthr = 1
I_max = 3 # (nA)
N = 100 # N 種類の入力電流
I = np.linspace(0, I_max, N) # 入力電流 (pA)
spikes = np.zeros((N, nt)) # スパイクの記録変数

for i in tqdm(range(N)):
    v = vreset; tlast = 0 # 初期化
    for t in range(nt):
        dv = (vrest - v + I[i]) / tc_m #　膜電位の導関数
        update = 1 if ((dt*t) > (tlast + tref)) else 0 # 不応期でないかの確認
        v = v + update*dv*dt # 膜電位の更新
        s = 1 if (v>=vthr) else 0 #発火時は 1, その他は 0 の出力
        tlast = tlast*(1-s) + dt*t*s # スパイク時刻の更新
        spikes[i, t] = s # 保存
        v = v*(1-s) + vreset*s # 膜電位のリセット

rate = np.sum(spikes, axis=1) / T # 発火率
plt.figure(figsize=(4, 3))
plt.plot(I, rate)
plt.xlabel('Input current (nA)'); plt.ylabel('Firing rate (Hz)')
plt.show()
```

#### 1.2.3.2 解析的F-I曲线描画



若存在一定且持续的电流$I$, 发火率(firing rate)的近似为：

$$rate\approx (\tau_mln\frac{R_mI}{R_mI-V_{th}})^{-1}$$

首先假设$t=t_1$时产生spiking, 此时膜电位reset, $V_m(t_1)=0$

[t,t_1]的膜电位根据LIF积分:

$$\tau\frac{dV_m(t)}{dt}=-V_m(t)+RI(t)\tag{10}$$ 

对(10)积分，有

$$\int_{t_1}^t\frac{\tau dV_m}{RI-V_m}=\int_{t_1}^t dt $$

$$ln(1-\frac{V_m(t)}{RI})=-\frac{t-t_1}{\tau}(因为V_m(t_1) =0)$$

因此有$$V_m(t)=RI[1-exp(-\frac{t-t_1}{\tau})]$$

假设$t_1$之后在$t_2$首次产生spking, 此时膜电位$V_m(t_2)=V_{th}$

将t=t_2带入上式，有


$$V_{th}=RI[1-exp(-\frac{t_2-t_1}{\tau})]$$

因此$T=t_2-t_1=\tau ln\frac{RI}{RI-V_{th}}$

由于发火率为$1/T$, 于是

$$rate\approx \frac{1}{T}=(\tau ln \frac{RI}{RI-V_{th}})^{-1} \tag{1.9}$$

若将不反应期记为$\tau_{ref}$, 发火率为$1/(T+\tau_{ref})$ 


接下来基于基于(1.9)描绘F-I曲线

```python
tc_m = 1e-2 # 膜時定数 (s)
R = 1 #膜抵抗
vthr = 1 # 閾値電位 (mV)
tref = 5e-3 # 不応期 (s)
I_max = 3 # 最大電流
I = np.arange(0, I_max, 0.01) #入力電流
rate = 1 / (tref + tc_m*np.log(R*I / (R*I - vthr)))
rate[np.isnan(rate)] = 0 # nan to 0

# 描画
plt.figure(figsize=(4, 3))
plt.plot(I, rate, color="k")
plt.xlabel('Input current (nA)'); plt.ylabel('Firing rate (Hz)')
plt.xlim(0, I_max)
plt.show()
```

### 1.3  Izhikevich/Simple模型

Izhikevich模型为Izhikevich在2003年考案的模型。HH模型虽然能够很好还原生理学模型，但是复杂且在数学上解析困难，计算量大，难以进行大规模的模拟。于是考虑更简单的能够再现生理学现象的发火模型。


Izhikevich模型则为其中一种模型。（下简略为$lz$模型）

lz模型使用仅有$2$变量的微分方程式。但可以模拟许多神经元活动。

$$C\frac{dv(t)}{dt}=k(v(t)-v_r)(v(t)-v_t)-u(t)+I$$

$$\frac{du(t)}{dt}=a\{b(v(t)-v_r)-u(t)\}$$

其中$v$与$u$为变量，$v$为膜电位(**membrane potential**, unit: mv)

$u$:回复电流(recorvery current, unit:pA)

$C$:膜容量（membrane capacitance, unit: pF）

$v_r$:静止膜电位 (resting membrane potential, unit: mv)

$v_t$:阈值电位（instantaneous threshold potential, unit: mv)

$a$:回复时间常数倒数(recovery time constant, unit: $ms^{-1}$) (**该数值越大，u回到原状的速度越快**)

$b$为$u$的对$v$的感受性(resonance, unit: pA/mV)

$k$: 关于神经元gain的常数.数值越小越容易发火。





lz模型的发火与HH较为详尽（与LIF不同）

lz模型的阈值为$v_t$,膜电位的reset不论是否超过阈值$v_t$，膜电位$v$成为peek电位$v_{peak}$时进行。

因此lz模型的阈值是变在膜电位的举动变化的点上，与阈值$v_t$存在差异。



此外，膜电位到达peak时, $u,v$以下列方式eset

$u\leftarrow u+d$

$v\leftarrow v_{reset}$

但是，$v_{reset}$是考虑了过分极的，相对于静止膜电位$v_r$小的值。

$d$ 代表尖峰点火期间激活的外向电流总净值，影响点火后的膜电位行为（单位 $pA$）。

```python
dt = 0.5; T = 500 # ms
nt = round(T/dt) # シミュレーションステップ数
# Regular spiking (RS) neurons
C = 100 # 膜容量 (pF)
a = 0.03 # 回復時定数の逆数 (1/ms)
b = -2 # u の v に対する共鳴度合い (pA/mV)
k = 0.7 # ゲイン (pA/mV)
d = 100 # 発火で活性化される正味の外向き電流 (pA)
vrest = -60 # 静止膜電位 (mV)
vreset = -50 # リセット電位 (mV)
vthr = -40 # 閾値電位 (mV)
vpeak = 35 #　ピーク電位 (mV)
t = np.arange(nt)*dt
I = 100*(t>50) - 100*(t>350) # 入力電流 (pA)
# 初期化 (膜電位, 膜電位 (t-1), 回復電流)
v = vrest; v_ = v; u = 0
v_arr = np.zeros(nt) # 膜電位を記録する配列
u_arr = np.zeros(nt) # 回復変数を記録する配列
# シミュレーション
for i in tqdm(range(nt)):
    dv = (k*(v - vrest)*(v - vthr) - u + I[i]) / C
    v = v + dt*dv # 膜電位の更新
    u = u + dt*(a*(b*(v_-vrest)-u)) # 膜電位の更新
    s = 1*(v>=vpeak) #発火時は 1, その他は 0 の出力
    u = u + d*s # 発火時に回復変数を上昇
    v = v*(1-s) + vreset*s # 発火時に膜電位をリセット
    v_ = v # v(t-1) <- v(t)
    v_arr[i] = v # 膜電位の値を保存
    u_arr[i] = u # 回復変数の値を保存


# 描画
t = np.arange(nt)*dt
plt.figure(figsize=(5, 5))
plt.subplot(2,1,1)
plt.plot(t, v_arr, color="k")
plt.ylabel('Membrane potential (mV)')
plt.subplot(2,1,2)
plt.plot(t, u_arr, color="k")
plt.xlabel('Time (ms)')
plt.ylabel('Recovery current (pA)')
plt.show()
```

图中上部分为膜电位$v$, 下部分为回复电流$u$

### 1.3.2 zhikevich模型类的実装

```python
class IzhikevichNeuron:
    def __init__(self, N, dt=0.5, C=250, a=0.01, b=-2,k=2.5, d=200, vrest=-60, vreset=-65, vthr=-20, vpeak=30):
        self.N = N
        self.dt = dt
        self.C = C

        self.a = a
        self.b = b
        self.d = d
        self.k = k
        self.vrest = vrest
        self.vreset = vreset
        self.vthr = vthr
        self.vpeak = vpeak
        self.u = np.zeros(N)
        self.v = self.vrest*np.ones(N)
        self.v_ = self.v
        
    def initialize_states(self, random_state=False):
        if random_state:
            self.v = self.vreset + np.random.rand(self.N) * (self.vthr-self.vreset)
        else:
            self.v = self.vrest*np.ones(self.N)
        self.u = np.zeros(self.N)
    
    def __call__(self, I):
        dv = (self.k*(self.v-self.vrest)*(self.v-self.vthr)-self.u+I)/self.C
        v = self.v + self.dt*dv
        u = self.u + self.dt*(self.a*(self.b*(self.v_-self.vrest)-self.u))
        
        s = 1*(v>=self.vpeak) #発火時は 1, その他は 0 の出力
        self.u = u + self.d*s
        self.v = v*(1-s) + self.vreset*s
        self.v_ = self.v
        return s
```

### 1.4 Inter-spike interval Model

目前为止讨论的模型均是基于电位随时间变化是否进行的发火模型。本节涉及的模型不考虑产生发火的过程，而是一个统计发火时间间隔(inter-spike interval, ISI)的现象论的模型。

ISI模型基于被称为**点过程(point process)**的统计模型，各个模型使用ISI遵从的分布名称命名。

使用频度较高的为 **Poisson process model**, 在poisson process model中考虑了不反应期的 **Poisson process with dead time(PPD)**模型

对于皮质发火，相比poisson过程更恰当的 **gamma process** model

在 SNN 中，ISI 模型主要用于在图像输入过程中将连续值编码为尖峰序列。




#### 1.4.1 Poisson Process Model

概率过程：随机变量随时间变化
在随机过程中，在连续时间轴上发生的一系列离散点事件称为点过程。


Poisson Process

时间$t$为止，产生的点数$N(t)$服从poisson分布。

若是服从强度为$\lambda$的poisson分布，时刻$t$为止事件产生$n$回的概率为：

$$P[N(t)=n]=\frac{(\lambda t)^n}{n!}e^{-\lambda t}$$


存在使用下列4条件定义的poisson process

(1) 时刻0点数为0

$P[N(0)=0]=1$

(2) $[t,t+\Delta t)$ 产生一点的概率

$P[N(t + Δt) -N(t) = 1] = \lambda(t)Δt + o(Δt)$

(3) 微小时间间隔$\Delta t$内不产生2个以上的点

$P[N(t + Δt) - N(t) = 2] = o(Δt)$

(4) 对于$t_1<t_2<..<t_n$

$N(t_2) - N(t_1);N(t_3) - N(t_2); ... ;N(t_n)- N(t_{n － 1})$ 相互独立

$N(t)$分布服从强度母数$\lambda(t)$的泊松分布:

強度函数(点为spike的情况, 发火率) が$\lambda(t) = \lambda$ (常数)

点间隔的概率变量为$T$, 服从$Exp(\lambda)=\lambda e^{-\lambda t}\;if\;t\ge 0\quad else\;0$

可从4条件以及Chapman-Kolmogorov式子求出。

也可从$P[N(t)=n]=\frac{(\lambda t)^n}{n!}e^{-\lambda t}$微分求出

$$F(t;\lambda)=P(T<t)=1-P(T>t)=1-P(N(t) = 0)=1-e^{-\lambda t}$$

对F微分

$$f(t;\lambda)=\frac{dF(t;\lambda)}{dt}=\lambda e^{-\lambda t}$$

##### 1.4.1.1 定常泊松过程 (homogeneous poisson process)

可利用ISI遵从指数分布或是2条件进行实现，后者计算量更少，但是为了之后的gamma过程的实现，这里使用前者。

```python
import numpy as np
dt = 1e-3; T = 1; nt = round(T/dt) # simulation时间
n_neurons = 10 # 神经元数
fr = 30 # poisson spike发火率(Hz)


isi = np.random.exponential(scale = 1/(fr*dt), size=(round(nt*1.5/fr), n_neurons)) #生成poission分布
spike_time = np.cumsum(isi, axis=0) # isi沿第二维度的前缀和

spike_time[spike_time > nt - 1] = 0 # nt を超える場合を0 に
spike_time = spike_time.astype(np.int32)

spikes = np.zeros((nt, n_neurons)) # spike记录变量
for i in range(n_neurons):
    spikes[spike_time[:, i], i] = 1
    spikes[0] = 0 # (spike_time=0) の発火を削除

print("Num. of spikes:", np.sum(spikes))
print("Firing rate:", np.sum(spikes)/(n_neurons*T))

# 自己猜isi生成了随机发火的间隔时间，服从指数分布。间隔时间累加后获得一个模拟的发火时间。
```

$np.random.exponential()$接受scale以及size. 其中scale是将概率密度函数看作$f(t;\frac{1}{\beta})$时的$\beta=1/\lambda$ (平均为$\beta$)

于是发火率为$fr\cdot dt$时，使用`np.random.exponential(1/(fr * dt))`

```python
# 描画
import matplotlib.pyplot as plt
t = np.arange(nt)*dt
plt.figure(figsize=(5, 4))
for i in range(n_neurons):
    plt.plot(t, spikes[:, i]*(i+1), 'ko', markersize=2)
plt.xlabel('Time (s)')
plt.ylabel('Neuron index')
plt.xlim(0, T); plt.ylim(0.5, n_neurons+0.5)
plt.show()
```

图表现了各个神经元点火的点。也被称为 **raster plot**

### 1.4.3 Gamma过程

Gamma过程为点时间分布遵从Gamma分布的过程

Gamma分布的概率密度函数:

$$f(t;k;\theta)=t^{k-1}\frac{e^{-t/\theta}}{\theta^k\Gamma(k)}$$

$$\Gamma(k)=\int_0^\infty x^{k-1}e^{-x}dx$$

平均:$k\theta$

发火率:$\lambda=1/k\theta$

```python
import scipy.special as sps
dt = 1e-3; T = 1; nt = round(T/dt) # シミュレーション時間
n_neurons = 10 # ニューロンの数
fr = 30 # ガンマスパイクの発火率(Hz)
k = 12 # k=1 のときはポアソン過程に一致
theta = 1/(k*(fr*dt)) # fr = 1/(k*theta)
isi = np.random.gamma(shape=k, scale=theta, size=(int(nt*1.5/fr), n_neurons))
spike_time = np.cumsum(isi, axis=0) # ISI を累積
spike_time[spike_time > nt - 1] = -1 # nt を超える場合を0 に
spike_time = spike_time.astype(np.int32) # float to int
spikes = np.zeros((nt, n_neurons)) # スパイク記録変数
for i in range(n_neurons):
    spikes[spike_time[:, i], i] = 1
spikes[0] = 0 # (spike_time=0) の発火を削除
print("Num. of spikes:", np.sum(spikes))
print("Firing rate:", np.sum(spikes)/(n_neurons*T))
```

```python
t = np.arange(nt)*dt
plt.figure(figsize=(5, 5))
plt.subplot(2,1,1) # ISI のヒストグラム
count, bins, ignored = plt.hist(isi.flatten(), 50, density=True, color="gray", alpha=0.5)
y = bins**(k-1)*(np.exp(-bins/theta) / (sps.gamma(k)*theta**k))
plt.plot(bins, y, linewidth=2, color="k")
plt.title('k=' + str(k))
plt.xlabel('ISI (ms)')
plt.ylabel('Probability density')


plt.subplot(2,1,2) # ラスタープロット
for i in range(n_neurons):
    plt.plot(t, spikes[:, i]*(i+1), 'ko', markersize=2)

plt.xlabel('Time (s)')
plt.ylabel('Neuron index')
plt.xlim(0, T)
plt.ylim(0.5, n_neurons+0.5)
plt.show()
```

### 1.5.1 Fring Rate Model

firing rate model不对神经活动的spike建模，而是对firing rate建模

令时间窗口大小为$T$, 发火回数为

$$ n = \int_0^T dt \delta(t-t_k)$$

其中$\delta$为Dirac常数，$t_k$为spike发生时刻

发火率$r=n/T$

#### 1.5.2 离散时间发火率模型

$\bold r = f(W\bold u+\bold b)$

$\bold u\in\mathbb R^{N_{pre}}$: 突触前细胞的发火率

$\bold r\in\mathbb R^{N_{post}}$: 突触后细胞的发火率

$\bold b\in\mathbb R^{N_{post}}$: bias

$W\in\mathbb R^{N_{post}\times N_{pre}}$: 突触强度矩阵 

$f(\cdot)$: 活性化函数。使用类似sigmoid的非线性函数


重要的事情是 **活性化関数表现了F-I curve**



#### 1.5.3 连续时间发火率模型 

基于Wilson-Cowan model的一种模型。


带有递归结合连续时间型RNN式子如下

$$\tau\frac{d\bold r}{dt}=-\bold r+f(W^{rec}\bold r+W^{in}\bold u+\bold b)$$

$\tau$为发火率的时间常数。

$\bold u\in\mathbb R^{N_{pre}}$: 突触前细胞的发火率

$\bold r\in\mathbb R^{N_{post}}$: 突触后细胞的发火率

$\bold b\in\mathbb R^{N_{post}}$: bias

$W\in\mathbb R^{N_{post}\times N_{pre}}$: 突触强度矩阵 

$f(\cdot)$: 活性化函数/发火率应答函数。使用类似sigmoid的非线性函数




##### **Euler 近似(first-order Euler approximation)**

该式子可以使用**Euler 近似(first-order Euler approximation)**

$$r_t = (1 - \alpha)\bold r_{t-1} + \alpha f(W^{rec}\bold r_{t-1} +W^{in}u_t + \bold b)$$

离散化后为被称为**leaky RNN**模型的一种

令$t$为层的数时，与ResNet形状类似。 

$(1 - \alpha)\bold r_{t-1}$为shortcut径路。

事实上，在这种形式下，捷径在 Backpropagation through time(BPTT) 中不易出现梯度损失和发散。(另一种门结构，如 LSTM 和 GRU 中的门结构 也能解决梯度丢失和发散问题）。

```python

import chainer
import chainer.functions as F
import chainer.links as L
from chainer import cuda
from chainer import Variable
xp = cuda.cupy
# import numpy as xp
class leakyRNN(chainer.Chain):
    def __init__(self, inp=32, mid=128, alpha=0.2, sigma_rec=0.1):
        super(leakyRNN, self).__init__()
        """
        Leaky RNN unit.
        Usage:
        >>> network = leakyRNN()
        >>> network.reset_state()
        >>> x = xp.ones((1, 32)).astype(xp.float32)
        >>> y = network(x)
        >>> y.shape
        (1, 128)
        """
        with self.init_scope():
            self.Wx = L.Linear(inp, mid) # feed-forward
            self.Wr = L.Linear(mid, mid, nobias=True) # recurrent
            self.inp = inp # 入力ユニット数
            self.mid = mid # 出力ユニット数
            self.alpha = alpha # tau / dt
            self.sigma_rec = sigma_rec # standard deviation of input noise

    def reset_state(self, r=None):
        self.r = r
    
    def initialize_state(self, shape):
        self.r = Variable(xp.zeros((shape[0], self.mid),
        dtype=self.xp.float32))
    
    def forward(self, x):
        if self.r is None:
            self.initialize_state(x.shape)
        z = self.Wr(self.r) + self.Wx(x)
        
        if self.sigma_rec is not None:
            z += xp.random.normal(0, self.sigma_rec,
        (x.shape[0], self.mid)) # Add noise
        r = (1 - self.alpha)*self.r + self.alpha*F.relu(z)
        self.r = r
        return r
```

#### 通过概率的突触电流噪声表现

有考虑脑内噪声(neuronal noise)的突触输入的情况。

例如 **LIF**考虑Gaussian噪声 ($\epsilon(t)\sim \mathcal N({\mu},{\sigma}^2)$)的情况

$$\tau_m\frac{dV_m(t)}{dt}=-(V_m(t)-V_{rest})+R_mI(t)+\epsilon(t)$$

像这样线性的$-V_m(t)$drift项与高斯无噪声项的某概率微分方程式(stochastic differential equations)形成的概率过程被称为 **Ornstein-Uhlenbeck(OU)** 过程

$\eta(t)$遵从$\mathcal N(0,1)$时，

$\epsilon(t)=\mu+\sigma\eta(t)$

如果噪声项 $\epsilon(t)$为遵循发火率的泊松过程，突触前细胞数为 $N_{pre}$，与第 $i$ 个突触的突触强度成比例的常数为 $J_i$，则 $\mu = ⟨J_i⟩N_{pre}$

$\sigma^2 = ⟨J_i^2⟩N_{pre}$。其中，⟨⟩ 表示取平均值。这就是**扩散近似(Diffusion approximation)**。

这是通过坎贝尔定理得到的。在对方程 1.31 进行离散化模拟时，必须小心谨慎。如果右侧只留下噪声项:

$$\tau_m\frac{dV_m(t)}{dt}=\epsilon(t)$$

进行欧拉离散化之后：

$$V_m(t + Δt) = V_m(t) +\frac{1}{\tau_m}\epsilon_1(t) \tag{(1.33)}$$

这种用欧拉方法对随机微分方程进行离散化的方法被称为**Euler-Maruyama法。**

对$\Delta$修正为$\Delta/2$, 也同样

$$V_m(t + Δt) = V_m(t+\Delta/2) +\frac{1}{\tau_m}\epsilon_1(t)=V_m(t)+\frac{1}{\tau_m}[\epsilon_1(t)+\epsilon_2(t)]\tag {(1.34)}$$

方程 (1.33) 和 (1.34) 中噪声项的标准偏差分别为 $\sigma/\tau_m$, $\sqrt 2\sigma/\tau_m$, 这意味着噪声项的标准偏差会随着时间步长的变化而变化，为了避免这种情况，我们考虑对噪声项进行$\sqrt{\Delta t}$乘法。


$$V_m(t + Δt) = V_m(t) +\frac{\sqrt{\Delta t}}{\tau_m}\epsilon_1(t)$$
